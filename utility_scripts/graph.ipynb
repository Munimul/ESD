```python
import numpy as np
import pandas as pd
```

```python
labels=pd.read_csv('C:/Pranto/Research/ESD/Data/Ready to Train/all_labelled/data/all/original/allLabels.csv')
```

```python
labels['width']
```

```python
import matplotlib.pyplot as plt
```

```python
rW=labels['width']/416
rH=labels['height']/416

```

```python
# Import seaborn
import seaborn as sns

# Apply the default theme

sns.set_style("white")
ax = sns.scatterplot(x=rW, y=rH )

ax.set(xlabel ="Relative Width", ylabel = "Relative Height", title ='Relative Width & Height of Bounding Boxes')

plt.savefig("C:/Pranto/Research/ESD/Data/Ready to Train/all_labelled/data/all/original/seaborn_plot.png")
# Create a visualization
```

```python
area=labels['width']*labels['height']/(416*416)
lx=sns.displot( x=area, kde=True)
lx.set(xlabel ="Relative area")
plt.savefig("C:/Pranto/Research/ESD/Data/Ready to Train/all_labelled/data/all/original/area_plot.png")
```
